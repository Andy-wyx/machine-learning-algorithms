```python
pop = dict(DNA=5 * np.random.rand(1, 1).repeat(100, axis=0),   
           mut_strength=np.random.rand(100, 1))  
print(pop)
```

```python
kids = {'DNA': np.empty((50, 1))}    # kids['DNA']:(50,1)
kids['mut_strength'] = np.empty_like(kids['DNA'])  #kids['mut_strength']:(50,1)
for kv, ks in zip(kids['DNA'], kids['mut_strength']):
    # crossover 
    p1, p2 = np.random.choice(np.arange(100), size=2, replace=False) # two different random numbers from 0~99
    print('p1',p1)
    print('p2',p2)
    
    cp = np.random.randint(0, 2, 1, dtype=np.bool_)  # crossover points    # true or false
    print('cp',cp)
    kv[cp] = pop['DNA'][p1, cp]         #if true:kv[1]=pop['DNA'][p1, 1]=[] if false: kv[0]=pop['DNA'][p1, 0]=
    print('1',kv[cp])
    kv[~cp] = pop['DNA'][p2, ~cp]       # if true: kv[-2]=pop['DNA'][p2, -2] if false: kv[-1]=pop['DNA'][p2, -1]  
    print('2',kv[~cp])
    ks[cp] = pop['mut_strength'][p1, cp]
    print('3',ks[cp])
    ks[~cp] = pop['mut_strength'][p2, ~cp]
    print('4',ks[~cp])
```
